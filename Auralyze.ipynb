```python
from pytubefix import YouTube
from pytubefix.cli import on_progress

url = input("Digite a url do youtube: ")

yt = YouTube(url, on_progress_callback=on_progress)
print(yt.title)

ys = yt.streams.get_audio_only()
ys.download(filename="audioyt.m4a")
```

```python
import whisper

# Install openai-whisper if not already installed
# !pip install openai-whisper


modelo = whisper.load_model("base")


resultado = modelo.transcribe("audioyt.m4a", fp16=False, language ="pt")


with open("transcricao.txt", "w", encoding="utf-8") as f:
    f.write(resultado["text"])
```

```python
from google import genai

client = genai.Client(api_key="API")

with open('transcricao.txt', 'r') as arquivo:
    conteudo = arquivo.read()
    pergunta = f"""
Faça um resumo breve do
contedudo: {conteudo}
"""

response = client.models.generate_content(
    model='gemini-flash-lite-latest',
    contents=pergunta
)

resp = response.text
print(resp)
```
